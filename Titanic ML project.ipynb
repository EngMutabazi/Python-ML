##  Group the data frame by values in Survived column, and count the number of occurrences of each group.

```python
survived_count = data_clean.groupby('survived')['survived'].count()
survived_count
```

```python
plt.figure(figsize=(4,5))
plt.bar(survived_count.index,survived_count.values)
plt.title("Grouped by survivals")
plt.xticks([0,1],['not survived','survived'])
for i, value in enumerate(survived_count.values):
    plt.text(i, value-5, str(value), fontsize=12, color='white',
             horizontalalignment='center', verticalalignment='center')
plt.show()
```

##  Group the data frame by classes in the pclass column, and count the number of occurrences of each group.

```python
group_pclass=data_clean.groupby('pclass')['pclass'].count()
```

```python
group_pclass
```

```python
plt.figure(figsize=(7,7))
plt.title('Grouped by pclass')
plt.pie(group_pclass.values, labels=['Class 1', 'Class 2', 'Class 3'], 
 autopct='%1.1f%%', textprops={'fontsize':13})
plt.show()
```

```python
group_sex_mf=data_clean.groupby('sex')['sex'].count()
```

```python
group_sex_mf
```

```python
plt.figure(figsize=(7,9))
plt.title('grouped by gender')
plt.pie(group_sex_mf.values,labels=['female','male'], autopct='%1.1f%%', textprops={'fontsize':10})
plt.show()
```
